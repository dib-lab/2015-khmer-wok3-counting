```python
import numpy
```

```python
def load_counts(filename):
    counts = {}
    for line in open(filename):
        line = line.strip().split(' ')
        name = line[0]
        ctl = [ float(c) for c in line[1:] ]
        counts[name] = ctl
    return counts


def build_array(mapping_filename, meds_filename):
    mapped = load_counts(mapping_filename)
    meds = load_counts(meds_filename)
    
    xx = []
    for k in meds:
        row = [mapped.get(k, [0])[0]]
        row.extend(meds[k])
        xx.append(row)
        
    return numpy.array(xx)
```

```python
# uncorrected RNAseq reads
rseq = build_array('rseq-mapped.counts', 'rseq-mapped.meds2')

# corrected mRNAseq reads (subset)
corr = build_array('rseq-mapped.counts', 'rseq-corr.meds2')
```

```python
#loglog(rseq[:,1], rseq[:,0], '.', label='path average')
loglog(rseq[:,2], rseq[:,0], 'g+', label='k-mer path average')

xlabel('path average in graph')
ylabel('average mapping count')
legend(loc='upper left')
title('k-mer counts alone')
savefig('fig1.png')
```

```python
loglog(rseq[:,2], rseq[:,0], 'g+', label='k-mer path average')
loglog(corr[:,2], corr[:,0], 'b.', label='using corrected reads')

xlabel('path average in graph')
ylabel('average mapping count')
legend(loc='upper left')
title('corrected mRNAseq vs uncorrected')
savefig('fig2.png')
```

```python
loglog(corr[:,2], corr[:,0], 'b.', label='corrected reads')
loglog(corr[:,3], corr[:,0], 'r*', label='aligning transcripts to graph')

#loglog(aa[:,0], aa[:,1], '.', label='aln')

#axis(xmin=1, xmax=1e6, ymin=1, ymax=1e6)
xlabel('path average in graph')
ylabel('average mapping count')
legend(loc='upper left')
savefig('fig3.png')
```
